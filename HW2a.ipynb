```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
# plot histogram of age
%matplotlib inline
train.Age.plot(kind='hist',bins=20)
```

```python
# plot x=Pclass, y=Fare
import matplotlib.pyplot as plt
plt.scatter(train['Pclass'], train['Fare'])
plt.xlabel('Pclass')
plt.ylabel('Fare')
```

```python
# crosstab of Survived and Pclass
pd.crosstab(train.Survived, train.Pclass)
```

```python
# assign survived = 1 to passengers of Pclass=1 and Age<10

test = pd.read_csv('http://bit.ly/kaggletest')
test.head()

# create new column
test['Survived'] = np.where((test.Pclass==1)&(test.Age<10), 1, 0)

pd.DataFrame({'PassengerId':test.PassengerId, 'Survived':test.Survived}).set_index('PassengerId').head(418)
```

```python
# write the DataFrame to a CSV file that can be submitted to Kaggle
pd.DataFrame({'PassengerId':test.PassengerId, 'Survived':test.Survived}).set_index('PassengerId').to_csv('sub.csv')
```
